```python
import numpy as np

def compute_agreement_metrics(expert, model):

    expert = np.array(expert)
    model = np.array(model)

    diff = model - expert
    mae = np.mean(np.abs(diff))
    rmse = np.sqrt(np.mean(diff ** 2))
    bias = np.mean(diff)
    print(f"MAE={mae:.2f}, RMSE={rmse:.2f}, Bias={bias:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_error_distribution(df, dimension_name, outfile='ErrorAnalysis/', figsize=(8,5), palette="Blues"):
    """
    Creates and saves a grouped bar chart of error-category percentages for each model.

    Parameters
    ----------
    df : pd.DataFrame
        Must include columns ['Model', 'rule'] for that dimension.
    dimension_name : str
        Name of the quality dimension (e.g., 'Syntactic', 'Semantic', 'Pragmatic').
    outfile : str, optional
        File path to save the figure (e.g., 'syntactic_errors.png').
        If None, the file is saved automatically using the dimension name.
    figsize : tuple
        Figure size in inches.
    palette : str
        Seaborn or Matplotlib palette name (default: 'Blues').
    """

    # --- Count number of occurrences per model per rule ---
    counts = df.groupby(['Model', 'rule']).size().reset_index(name='Count')

    # --- Pivot: rules as index, models as columns ---
    pivot_df = counts.pivot(index='rule', columns='Model', values='Count').fillna(0)

    #print(pivot_df)
    # --- Convert to percentages (normalize per model) ---
    #pivot_df = pivot_df.div(pivot_df.sum(axis=0), axis=1) * 100

    # --- Sort categories alphabetically (optional) ---
    pivot_df = pivot_df.reindex(sorted(pivot_df.index))

    # --- Define consistent blues palette ---
    colors = sns.color_palette(palette, n_colors=len(pivot_df.columns))

    # --- Plot grouped bars ---
    ax = pivot_df.plot(kind='bar', figsize=figsize, width=0.8, color=colors)

    # --- Labels & style ---
    plt.ylabel('Percentage of detected errors (%)')
    plt.xlabel('')
    plt.title(f'Error category distribution across models ({dimension_name})')
    plt.xticks(rotation=30, ha='right')
    plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()

    # --- Save figure ---
    if outfile is None:
        outfile = f"{dimension_name.lower()}_error_distribution.png"
    outfile = outfile + f"{dimension_name.lower()}_error_distribution.png"
    plt.savefig(outfile, dpi=300)
    plt.show()

    plt.close()   # closes figure (no inline display)

    print(f"✅ Saved: {outfile}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_violin_three_metrics(syntactic, semantic, pragmatic,  save_path=None ,title=None):
    """
    Plot violin + boxplots for three error metrics:
    Syntactic, Semantic, Pragmatic.

    Parameters
    ----------
    syntactic : pandas.Series
        Series of syntactic error counts.
    semantic : pandas.Series
        Series of semantic error counts.
    pragmatic : pandas.Series
        Series of pragmatic error counts.
    title : str, optional
        Optional title for the figure.
    save_path : str, optional
        Path to save figure (e.g. "plot.png"), or None to just show it.
    """

    # Build long-format data
    metrics = ["Syntactic Errors", "Semantic Errors", "Pragmatic Errors"]
    values = [syntactic, semantic, pragmatic]

    records = []
    for metric, series in zip(metrics, values):
        for v in series.dropna():
            records.append({"Metric": metric, "Value": v})

    long_data = pd.DataFrame.from_records(records)

    # ---- Plot ----
    sns.set(style="whitegrid", font_scale=1.1)
    fig, ax = plt.subplots(figsize=(10, 6))

    ax.set_ylim(0, 6)
    ax.set_yticks(np.arange(0, 7, 1))

    sns.violinplot(
        x="Metric",
        y="Value",
        data=long_data,
        palette="Blues",
        inner=None,
        cut=0,
        linewidth=1.5,
        density_norm='width',
        bw_adjust=0.7,
        ax=ax
    )



    # Add centered white boxplots
    x_positions = range(len(metrics))
    width = 0.15

    for i, metric in enumerate(metrics):
        vals = long_data[long_data["Metric"] == metric]["Value"]
        ax.boxplot(
            vals,
            positions=[i],
            widths=width,
            patch_artist=True,
            boxprops={'facecolor': 'white', 'edgecolor': 'black', 'linewidth': 2},
            whiskerprops={'color': 'black', 'linewidth': 2},
            capprops={'color': 'black', 'linewidth': 2},
            medianprops={'color': 'black', 'linewidth': 2},
            flierprops={'marker': 'o', 'markersize': 5, 'markerfacecolor': 'black', 'alpha': 0.5}
        )

    # Formatting
    max_val = long_data["Value"].max()
    step = max(1, int(np.ceil(max_val / 6)))
    ax.set_yticks(np.arange(0, max_val + step, step))

    ax.set_xlabel("Metric")
    ax.set_ylabel("Number of Errors")
    ax.set_xticks(range(len(metrics)))
    ax.set_xticklabels(metrics)

    if title:
        ax.set_title(title)

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=600, bbox_inches='tight')

    plt.show()
    return fig, ax
```

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Load the CSV files
llm_df = pd.read_csv('AnalysisGPTasEvaluator.csv')
human_df = pd.read_csv('AnalysisWithNewMethod.csv')



# Optional: Align rows by an ID column (e.g., 'DiagramID') if necessary
# merged_df = pd.merge(human_df, llm_df, on='DiagramID', suffixes=('_human', '_llm'))

# If aligned manually, assume same order
dimensions = ["SyntacticHuman",	"SyntacticLLM",	"SemanticHuman",	"SemanticLLM",	"PragmaticHuman",	"PragmaticLLM"]
correlations = []

for dim in dimensions:
    human_col = human_df[dim]
    llm_col = llm_df[dim]

    pearson_corr, pearson_p = pearsonr(human_col, llm_col)
    spearman_corr, spearman_p = spearmanr(human_col, llm_col)

    correlations.append({
        "Dimension": dim,
        "Pearson_r": round(pearson_corr, 3),
        "Pearson_p": round(pearson_p, 4),
        "Spearman_rho": round(spearman_corr, 3),
        "Spearman_p": round(spearman_p, 4)
    })

# Display results
correlation_df = pd.DataFrame(correlations)
print("\nCorrelation between Human and LLM Evaluations:\n")
print(correlation_df.to_string(index=False))
```

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is already loaded as `data`
# and contains the following columns:
# SynLLM, SynHuman, ClaudeSynLLM, CludeSynHuman, ClaudeCorrectSynLLM, ClaudeCorrectSynHuman

# Example for syntactic errors
#data = pd.read_csv('RQ2SynAnalysis.csv')
#data = pd.read_csv('RQ2SemAnalysis.csv')
data = pd.read_csv('RQ2PragAnalysis.csv')
```

```python

SynLLM= data['SynLLM']
SynHuman= data['SynHuman']
ClaudeSynLLM= data['ClaudeSynLLM']
CludeSynHuman= data['CludeSynHuman']
ClaudeCorrectSynLLM= data['ClaudeCorrectSynLLM']
ClaudeCorrectSynHuman= data['ClaudeCorrectSynHuman']
```

```python
def analysis(total_correct,total_reported,total_ground_truth,item):
  precision = total_correct / total_reported if total_reported > 0 else 0
  recall = total_correct / total_ground_truth if total_ground_truth > 0 else 0
  f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
  # Print evaluation metrics
  print("Syntactic Quality Evaluation Metrics:"+item)
  print(f"Precision: {precision:.2f}")
  print(f"Recall: {recall:.2f}")
  print(f"F1 Score: {f1:.2f}")
```

```python
# LLM diagrams
total_correct = data['ClaudeCorrectSynLLM'].sum()
total_reported= data['ClaudeSynLLM'].sum()
total_ground_truth = data['SynLLM'].sum()
analysis(total_correct,total_reported,total_ground_truth, 'LLM')

# Step 3: Correlation analysis
pearson_corr, _ = pearsonr(ClaudeSynLLM, SynLLM)
spearman_corr, _ = spearmanr(ClaudeSynLLM, SynLLM)


print(f"Pearson Correlation: {pearson_corr:.2f}")
print(f"Spearman Correlation: {spearman_corr:.2f}")


# Human diagrams
total_correct = data['ClaudeCorrectSynHuman'].sum()
total_reported= data['CludeSynHuman'].sum()
total_ground_truth = data['SynHuman'].sum()
analysis(total_correct,total_reported,total_ground_truth, 'Human')

# Step 3: Correlation analysis
pearson_corr, _ = pearsonr(CludeSynHuman, SynHuman)
spearman_corr, _ = spearmanr(CludeSynHuman, SynHuman)


print(f"Pearson Correlation: {pearson_corr:.2f}")
print(f"Spearman Correlation: {spearman_corr:.2f}")



# Total diagrams
total_correct = data['ClaudeCorrectSynHuman'].sum() + data['ClaudeCorrectSynLLM'].sum()
total_reported= data['CludeSynHuman'].sum() + data['ClaudeSynLLM'].sum()
total_ground_truth = data['SynHuman'].sum() + data['SynLLM'].sum()

analysis(total_correct,total_reported,total_ground_truth, 'Total')

# Step 3: Correlation analysis
ground_truth = pd.concat([data['SynLLM'], data['SynHuman']], ignore_index=True)
claude_reported = pd.concat([data['ClaudeSynLLM'], data['CludeSynHuman']], ignore_index=True)

pearson_corr, _ = pearsonr(claude_reported, ground_truth)
spearman_corr, _ = spearmanr(claude_reported, ground_truth)


print(f"Pearson Correlation: {pearson_corr:.2f}")
print(f"Spearman Correlation: {spearman_corr:.2f}")

```

```python
corr_df = pd.DataFrame({
    'GroundTruth': ground_truth,
    'ClaudeReported': claude_reported
})
```

```python

# Step 4: Scatter plot of Claude vs. Ground Truth counts
plt.figure(figsize=(6, 5))
sns.regplot(x='GroundTruth', y='ClaudeReported', data=corr_df)
plt.title('Pragmatic Error Count Correlation\nLLM generated vs. Ground Truth')
plt.xlabel('Ground Truth Error Count')
plt.ylabel('LLM Generated Error Count')
plt.grid(True)
plt.tight_layout()

plt.savefig("pragScatter.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate totals
gt_llm = data['SynLLM'].sum()
gt_human = data['SynHuman'].sum()
gt_total = gt_llm + gt_human

detected_llm = data['ClaudeSynLLM'].sum()
detected_human = data['CludeSynHuman'].sum()
detected_total = detected_llm + detected_human

correct_llm = data['ClaudeCorrectSynLLM'].sum()
correct_human = data['ClaudeCorrectSynHuman'].sum()
correct_total = correct_llm + correct_human

# Create DataFrame
plot_df = pd.DataFrame({
    'Source': ['LLM-generated', 'Human-generated', 'Total'],
    'Ground Truth': [gt_llm, gt_human, gt_total],
    'LLM-Reported': [detected_llm, detected_human, detected_total],
    'Ground Truth Matches': [correct_llm, correct_human, correct_total]
})

# Melt for seaborn
plot_df_melted = plot_df.melt(id_vars='Source', var_name='Category', value_name='Count')

# Custom color palette (academic-friendly blue tones)
palette = {
    'Ground Truth': '#003f5c',
    'LLM-Reported': '#7a5195',
    'Ground Truth Matches': '#2f4b7c'
}

# Plot
plt.figure(figsize=(6, 5))
sns.barplot(data=plot_df_melted, x='Source', y='Count', hue='Category', palette=palette, dodge=True)
plt.title('Pragmatic Error Counts by Diagram Source', fontsize=10)
plt.ylabel('Number of Issues')
plt.xlabel('Diagram Source')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Metric', loc='upper left', fontsize=9, title_fontsize=10)
sns.despine()
plt.tight_layout()
plt.savefig("spragBar.png", dpi=600, bbox_inches='tight')
plt.show()
```
